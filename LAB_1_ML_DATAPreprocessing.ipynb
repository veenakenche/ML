```python
diabetes = pd.read_csv("diabetes.csv")
```

```python
print(diabetes.columns)
```

```python
diabetes.columns = diabetes.columns.str.strip()  # removes extra spaces
print(diabetes.columns)
```

```python
cols_with_zero_as_missing = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in cols_with_zero_as_missing:
    if col in diabetes.columns:
        diabetes[col] = diabetes[col].replace(0, np.nan)
        diabetes[col].fillna(diabetes[col].median(), inplace=True)
```

```python
print(diabetes.columns)
```

```python
print(diabetes.columns.tolist())
```

```python
import numpy as np

# Replace 0 with NaN for numeric columns only
num_cols = diabetes.select_dtypes(include=np.number).columns

diabetes[num_cols] = diabetes[num_cols].replace(0, np.nan)
diabetes[num_cols] = diabetes[num_cols].fillna(diabetes[num_cols].median())
```

```python
Q1 = diabetes[num_cols].quantile(0.25)
Q3 = diabetes[num_cols].quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~((diabetes[num_cols] < (Q1 - 1.5 * IQR)) |
                      (diabetes[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

minmax = MinMaxScaler()
standard = StandardScaler()

diabetes_minmax = pd.DataFrame(minmax.fit_transform(diabetes[num_cols]), columns=num_cols)
diabetes_standard = pd.DataFrame(standard.fit_transform(diabetes[num_cols]), columns=num_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in diabetes.select_dtypes(include='object').columns:
    diabetes[col] = le.fit_transform(diabetes[col])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
diabetes_minmax = pd.DataFrame(scaler_minmax.fit_transform(diabetes),
                               columns=diabetes.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler_std = StandardScaler()
diabetes_standard = pd.DataFrame(scaler_std.fit_transform(diabetes),
                                 columns=diabetes.columns)
```
