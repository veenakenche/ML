```python
adult = pd.read_csv("adult.csv")
adult.columns = adult.columns.str.strip()

print(adult.head())
print(adult.info())
```

```python
adult.replace(" ?", np.nan, inplace=True)
adult.fillna(adult.mode().iloc[0], inplace=True)
```

```python
le = LabelEncoder()
for col in adult.select_dtypes(include='object').columns:
    adult[col] = le.fit_transform(adult[col])
```

```python
num_cols_adult = adult.select_dtypes(include=np.number).columns

Q1 = adult[num_cols_adult].quantile(0.25)
Q3 = adult[num_cols_adult].quantile(0.75)
IQR = Q3 - Q1

adult = adult[~((adult[num_cols_adult] < (Q1 - 1.5 * IQR)) |
                (adult[num_cols_adult] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
adult_minmax = pd.DataFrame(MinMaxScaler().fit_transform(adult),
                            columns=adult.columns)
```

```python
adult_standard = pd.DataFrame(StandardScaler().fit_transform(adult),
                              columns=adult.columns)
```

```python
print("First 5 rows:\n", adult.head())
print("\nLast 5 rows:\n", adult.tail())
print("\nShape:", adult.shape)
print("\nColumns:", adult.columns.tolist())
print("\nData Types:\n", adult.dtypes)
```

```python
for col in adult.select_dtypes(include='object').columns:
    print(f"\nUnique values in {col}:\n", adult[col].unique())
```
